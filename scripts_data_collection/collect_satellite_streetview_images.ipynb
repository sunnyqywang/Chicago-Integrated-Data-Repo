### Aerial images

```python
import googlemaps as gmap
```

```python
API_key = 'Your API Key'
my_gmap = gmap.Client(key = API_key)
# dir(googlemaps.maps)
```

```python
# Yeah. It works.
# Check: https://github.com/googlemaps/google-maps-services-python/blob/master/googlemaps/maps.py
# Check: https://developers.google.com/maps/documentation/maps-static/overview

f = open('../data/image/chicago_2.jpg', 'wb')
for chunk in my_gmap.static_map(size=(400,400), 
                                center=(41.85896,-87.653719), 
                                zoom=12, 
                                maptype='roadmap'):
    # this is an example in NYC. Zoom = 14 seems to fit the urban areas. 
    # Q: How to precisely control the spatial size of satellite images?
    if chunk:
        f.write(chunk)

f.close()
```
